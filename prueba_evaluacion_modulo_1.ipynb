```python
import re

class TiendaOnLine: 

    def __init__(self):
        
        self.inventario = []
        self.clientes = {}
        self.ventas_totales = 0

#1. Definimos la clase con sus atributos --> inventario, clientes y ventas totales.
#   En este caso, tanto el inventario como la clientela, debían ir en una lista y en un diccionario vacíos respectivamente;
#   y el resgistro de las ventas totales comienza en 0. 
#   En este ejemplo, no es necesario añadir los atributos como parámetros (variables locales) debido a que estos no son valores (argumentos) específicos para instanciar la clase (crear el objeto).

    def agregar_producto(self, nombre, precio, cantidad):

        for producto in self.inventario:
            if producto == nombre:
                producto ["cantidad"] += cantidad # importante habilitar el espacio de 'cantidad'--> key ya que producto es un diccionario.
        else: #pertenece al for, no al if
            self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
            return #en caso de que no se cumpla el if, se ejecutará el else y el return llamará a la función.

#2. En este primer método, debemos saber si en el inventario existe un producto específico, por eso utilizamos for (iterar):
#   - Si el producto ya existía en el inventario --> se añade la cantidad.
#   - Si el producto no existía en el inventario --> se añade como un producto nuevo.
#   Aquí sí es necesario añadir los atributos debido a que son valores específicos para crear el objeto.

    def ver_inventario(self):

        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']}, precio: {producto['precio']} €, cantidad: {producto['cantidad']}") #se accede a clave-valor a través de la clave del diccionario ([""]).

#3. En este método, debemos ver qué productos (nombre, precio y cantidad) hay en el inventario, por eso utilizados nuevamente for (iterar):

    def buscar_producto(self, nombre):
        
        for producto in self.inventario:
            if producto ["nombre"] == nombre: #habilitar el espacio de nombre (clave).
                print(f"Nombre: {producto['nombre']}, precio: {producto['precio']} €, cantidad: {producto['cantidad']}")
                return
        else:
            print(f"Este producto no se encuentra en el inventario")
            

#4. En éste debemos buscar qué producto hay en el inventario a través del nombre (iterar con for) y printear nombre, precio y cantidad.
#   También añadirmos un atributo como parámetro ya que es específico.

    def actualizar_stock(self, nombre, cantidad):
    
        for producto in self.inventario:
            if producto ["nombre"] == nombre:
                producto["cantidad"] += cantidad #'cantidad'--> key
                return
        else:
            print(f"{producto ['nombre']} no se encuentra en el inventario.") #comillas dobles y simples.

#5. En éste debemos actualizar el producto del inventario añaniendo cantidad.

    def eliminar_producto(self, nombre):

        for ind, producto in enumerate(self.inventario):
            if producto["nombre"] == nombre:
                self.inventario.pop(ind)
                return
        else:
            print(f"Ese producto {producto['nombre']} no se encuentra en el inventario.")

#6. En este método eliminamos un producto del inventario.
#   Utilizamos enumerate para poder saber el indice del producto cuando itera por el inventario (lista)
#   Si el nombre del producto elegido se encuentra en la lista, es eliminado del ésta utilizando pop().

    def calcular_valor_inventario(self):
        
        valor_inventario = 0
        for producto in self.inventario:
            valor_inventario += producto["precio"] * producto["cantidad"]
        return valor_inventario
    
#7. En este método calculamos el valor total del inventario.
#   Cuando iteramos y encontramos el producto en el inventario, se añade al valor del inventario una multiplicación:
#   ---> precio del producto * cantidad del producto = total del valor <---


    def buscar_producto_regex(self, patron):
     
        productos_encontrados = [] # nueva variable para agregar los productos que han sido encontrados en el inventario.
        
        for producto in self.inventario:
            
            if re.search(patron, producto["nombre"]):  # si el nombre del producto buscado se encuentra en el inventario:
              productos_encontrados.append(producto)   # se añade a la lista creada de "productos_encontrados".
              print(f"Nombre: {producto['nombre']}, precio: {producto['precio']} €, cantidad: {producto['cantidad']}") # se printea el producto encontrado
                                                                                                                       # con todas sus caraterísticas.
            
        else:
            pass                                                                                                                          
         

#8. En este método buscamos un producto en el inventario existente. 
#   Si el producto buscado coincide con el patrón marcado, dicho producto se añadirá a la nueva lista "productos_encontrados". 
#   Utilizamos re.search ya que busca en toda la cadena y devolverá el objeto coincidente.
#   Estructura: --> re.search (patron_a_buscar, cadena_original).


    def agregar_cliente(self, nombre, email):
        
        if nombre not in self.clientes:
            self.clientes[nombre] = {"email": email, "compra": []}

        else:
            print(f"El cliente con nombre: {nombre} ya está registrado")
            return
            
    def ver_clientes(self):

        for nombre, dicc in self.clientes.items(): #para iterar por clave-valor del diccionario (dicc)
            if nombre in self.clientes:
                email= dicc["email"] #creamos la variable "email" ya que lo que buscamos es que del diccionario "dicc" nos saque el valor "email"y se agine a dicha variable 
                print(f"Nombre: {nombre}, email: {email}")
                
            else:
                print(f"El nombre {nombre} no está en el registro de clientes.")           
```

```python
clothe = TiendaOnLine()

clothe.agregar_producto("bufanda", 10, 20)
clothe.agregar_producto("mocasines", 60, 10)
clothe.agregar_producto("gorro", 5, 15)
clothe.agregar_producto("gabardina", 40, 12)

print("_______________________________________________________________________________________________")
print("INVENTARIO")
print(" ")
clothe.ver_inventario()

print("_______________________________________________________________________________________________")
print("BÚSQUEDA DEL PRODUCTO: mocasines")
print(" ")
clothe.buscar_producto("mocasines")

print("_______________________________________________________________________________________________")
print("ACTUALIZAR STOCK --> (gabardina, 10) <--- El inventario queda de la siguiente manera:")
print(" ")
clothe.actualizar_stock("gabardina", 10)
clothe.ver_inventario()

print("________________________________________________________________________________________________")
print("ELIMINAR PRODUCTO: bufanda. El inventario queda de la siguiente manera:")
print(" ")
clothe.eliminar_producto("bufanda")
clothe.ver_inventario()

print("_______________________________________________________________________________________________")
print("VALOR TOTAL DEL INVENTARIO")
print(" ")
print(f"El valor total que tiene el inventario la tienda online 'Clothe' es de: {clothe.calcular_valor_inventario()} €.")

print("________________________________________________________________________________________________")
print("PRODUCTO QUE SE ENCUENTRA EN EL INVENTARIO")
print(" ")
clothe.buscar_producto_regex("mocasines")
clothe.buscar_producto_regex("gorro")
clothe.buscar_producto_regex("gabardina")

print("_______________________________________________________________________________________________")
print("REGISTRO CLIENTELA")
print(" ")
clothe.agregar_cliente("Carolina", "carol.ina37@gmail.com")
clothe.agregar_cliente("Vera", "v230891.carriz@hotmail.com")
clothe.ver_clientes()
```
